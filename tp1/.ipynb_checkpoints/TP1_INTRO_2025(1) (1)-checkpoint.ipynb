Escriba entonces un grafico que permita entender que es lo que hará su codigo (pseudocodigo).

3) Existe una función/operación módulo (o resto). Piense si le puede ser util para el código.

4) Recuerde que para recorrer elementos en orden se usan listas.

5) Una vez tenga alguna primera versión del programa funcionando:
    5a) Analice la performance en tiempo (cuanto tarda su programa?). Use graficos de tiempo vs número de predicciones.
    5b) Evalue como mejorar la performance. Haga un pseudocodigo, y si encuentra como, implementelo. Haga un grafico de tiempo
            para comparar con 5a.

````python
def obtener_primos():
    res = []  # Lista para guardar los números primos
    for i in range(2, 21):  # Comienza desde 2, ya que 1 no es primo
        divisores = 0  # Contador de divisores
        for j in range(1, i + 1):  # Verifica los divisores de i
            if i % j == 0:
                divisores += 1
        if divisores == 2:  # Si tiene solo dos divisores, es primo
            res.append(i)
    return res
````

### Ejercicio 3:
- Selecciona o crea un archivo de texto y cuenta cuántas veces aparece una palabra dada (ingresada por el usuario).

- Cuente la frecuencia de cada palabra y luego represente el conteo en una nube de palabras (puede explorar la librería WordCloud o alguna que considere útil).

- Pruebe el código con algún archivo de texto, ¿para que podria usar esa representacion ? ¿le parece necesario excluir alguna palabra para realizar el gráfico? ¿Considera que está usted sesgado en la utilización de alguna muletilla/palabra al escribir?

````python
#Creamos una funcion que lee un texto, convierte mayusculas en minusculas y elimina caracteres de puntuacion.
import string
def procesar_texto(ruta_archivo):
    f =  open(ruta_archivo, 'r')
    texto = f.read()
    texto = texto.lower() #No discriminamos entre mayuscula/minuscula
    puntuacion = str.maketrans(string.punctuation,' ' * len(string.punctuation) ) # Dos traductores: puntuacion elimina ')', ';', etc
    separador_lineas = str.maketrans('\n', " " )        #separador_lineas elimina  '\n'

    texto =  texto.translate(separador_lineas)
    texto =  texto.translate(puntuacion)
    return(texto)
````

````python
#Conteo de una palabra en un texto
#El texto utilizado es: https://drive.google.com/file/d/1rvd8N6EqcQpG4r_dLXD_k9RxNSlHKmFP/view?usp=sharing
ruta_archivo = 'texto.txt'
def contar(palabra, ruta_archivo):
    contador = 0
    texto = procesar_texto(ruta_archivo)
    lista_palabras = texto.split()
    for pal in lista_palabras:
      if (pal == palabra):
        contador = contador +1
    return (contador)

palabra = input('Ingresar palabra a buscar ')
contar(palabra,ruta_archivo)
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
ruta_archivo = 'texto.txt'
texto = procesar_texto(ruta_archivo)
lista_palabras = texto.split()
#Creamos un diccionario que contiene como clave a las palabras del texto y significado su frecuencia.
diccionario = {}
for palabra in lista_palabras:
  if( not (palabra in diccionario)) :
    diccionario.update({f"{palabra}": 1})
  else: diccionario[palabra] += 1
#Representamos las frecuencias con una nube de palabras
wordcloud = WordCloud(background_color='white').generate_from_frequencies(diccionario)
plt.imshow(wordcloud)


````
